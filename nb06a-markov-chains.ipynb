# `nb06a`: Markov chains

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
```

```python
# 1d Gaussian random walk
def gaussian_random_walk_1d(N, T, mu=0, sigma=1):
    x = np.zeros((N, T))
    x[:, 0] = np.random.normal(mu, sigma, size=(N,))
    for t in range(1, T):
        step = np.random.normal(mu, sigma, size=(N,))
        x[:, t] = x[:, t-1] + step
    return x

fig = plt.figure(figsize=(6.4, 6.4), dpi=200)
ax = plt.axes()

x = gaussian_random_walk_1d(N=20, T=100, mu=0, sigma=1)
for i in range(x.shape[0]):
    ax.plot(x[i], alpha=0.6)

ax.set_xlabel('t')
ax.set_ylabel(r'$x_t$')

plt.savefig("gaussian_random_walk_1d.png", dpi=200)
plt.show()
```

```python
# Variance over time
x = gaussian_random_walk_1d(N=10000, T=1000, mu=0, sigma=1)
var = np.var(x, axis=0)
    
fig = plt.figure(figsize=(6.4, 6.4), dpi=200)
ax = plt.axes()
ax.plot(var, label=r'Gaussian random walk')
ax.set_xlabel('t')
ax.set_ylabel(r'$\mathbb{V}[x_t]$')
ax.legend()

plt.savefig("gaussian_random_walk_1d_variance.png", dpi=200)
plt.show()
```

```python
# 1d Gaussian random walk with drift
def gaussian_random_walk_1d_drift(N, T, mu=0, sigma=1, kappa=0.1):
    x = np.zeros((N, T))
    x[:, 0] = np.random.normal(mu, sigma, size=(N,))
    for t in range(1, T):
        step = np.random.normal(mu, sigma, size=(N,))
        x[:, t] = x[:, t-1] - kappa * (x[:, t-1] - 0.) + step
    return x

fig = plt.figure(figsize=(6.4, 6.4), dpi=200)
ax = plt.axes()

x = gaussian_random_walk_1d_drift(N=20, T=100, mu=0, sigma=1, kappa=0.1)
for i in range(x.shape[0]):
    ax.plot(x[i], alpha=0.6)

ax.set_xlabel('t')
ax.set_ylabel(r'$x_t$')

plt.savefig("gaussian_random_walk_1d_drift.png", dpi=200)
plt.show()
```

```python
# Variance over time
x = gaussian_random_walk_1d(N=10000, T=1000, mu=0, sigma=1)
x2 = gaussian_random_walk_1d_drift(N=10000, T=1000, mu=0, sigma=1, kappa=0.1)
var = np.var(x, axis=0)
var_drift = np.var(x2, axis=0)

fig = plt.figure(figsize=(6.4, 6.4), dpi=200)
ax = plt.axes()
ax.plot(var, label=r'Gaussian random walk')
ax.plot(var_drift, label=r'Gaussian random walk with drift')
ax.set_xlabel('t')
ax.set_ylabel(r'$\mathbb{V}[x_t]$')
ax.legend()

plt.savefig("gaussian_random_walk_variance_comparison.png", dpi=200)
plt.show()
```
